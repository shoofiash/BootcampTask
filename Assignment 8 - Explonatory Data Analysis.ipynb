```python
DataCleanedPrice = []
for i in range(len(house_price)):
  if (house_price["Currency"].iloc[i] == "USD"):
    DataCleanedPrice.append(house_price["Annual_Price"].iloc[i] * 15000)
  else:
    DataCleanedPrice.append(house_price["Annual_Price"].iloc[i])

house_price.insert(len(house_price.columns), "CleanedPrice", DataCleanedPrice)
```

### Nomor 5:

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
#Menampilkan matrix korelasi
new_house_price.corr()
```

```python
#Menampilkan visualisasi
plt.figure(figsize=(10,10))
plt.title("Matrix Korelasi Data")
sns.heatmap(new_house_price[["Area","No_Rooms","Total_Facilities","CleanedPrice"]].corr(), annot=True, linewidths=3)
plt.show()
```

```python
#Train Split
x = new_house_price[["Area","No_Rooms","Total_Facilities"]]
y = new_house_price["CleanedPrice"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
pd.DataFrame(x_train)
```

```python
pd.DataFrame(y_train)
```

```python
#Train a Model on Training Set
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Test The Model on Test Set
y_prediction = model.predict(x_test)
y_prediction
```

```python
y_test
```

```python
ax = sns.scatterplot(y_prediction, y_test)
ax.set_xlabel("Prediction")
ax.set_ylabel("Actual Test Data")
plt.show()
```

```python
#Evaluate Performance
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print('TEST SET EVALUATION')
print('===================')
print(f'Mean Absolute Error: {mean_absolute_error(y_test, y_prediction)}')
print(f'Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test, y_prediction))}')
print(f'R Squared Score: {r2_score(y_test, y_prediction)}')
```

```python
training_prediction = model.predict(x_train)
print('TRAIN SET EVALUATION')
print('===================')
print(f'Mean Absolute Error: {mean_absolute_error(y_train, training_prediction)}')
print(f'Root Mean Squared Error: {np.sqrt(mean_squared_error(y_train, training_prediction))}')
print(f'R Squared Score: {r2_score(y_train, training_prediction)}')
```

Hasil antara test dan train cukup berdekatan. Tetapi hasil MAE serta RMSE sangat besar. MAE menunjukkan nilai kesalahan rata-rata yang error dari nilai sebenarnya dengan nilai prediksi. Sedangkan RMSE merupakan jumlah dari kesalahan kuadrat antara nilai sebenarnya dengan nilai prediksi yang telah ditentukan. Sehingga, dapat disimpulkan bahwa model yang dibuat merupakan model yang jelek karena memiliki nilai eror yang sangat besar. Hal ini bisa disebabkan karena data tidak terdistribusi dengan normal. Pada hasil scatter, terlihat bahwa grafik tidak menciptakan tren. Sehingga, model ini tidak baik dalam memprediksi hubungan antara area, jumlah kamar, serta total fasilitas dengan harga.
